```python
import csv
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Read the data from the CSV file
gdp = pd.read_csv('data_all_teams_gdp.csv')

# Create a new figure
plt.figure()

# Plot the data
plt.plot(gdp['year'], gdp['gdp'], label='GDP')

```

```python
# Read the data from the CSV file into a DataFrame
data_df = pd.read_csv('data_all_teams.csv')

# Plot the data
plt.figure(figsize=(10, 6))
data_df.groupby('profession')['income'].mean().plot(kind='bar')
plt.xlabel('Profession')
plt.ylabel('Average Income')
plt.title('Average Income by Profession')
plt.show()
```

```python
# Group by year and count the number of citizens
citizen_population = data_df.groupby('year').size()

# Plot the citizen population for every year
plt.figure(figsize=(12, 6))
citizen_population.plot(kind='line')
plt.xlabel('Year')
plt.ylabel('Citizen Population')
plt.title('Citizen Population Over the Years')
plt.show()

# Additional insights: Average income by year
average_income_by_year = data_df.groupby('year')['income'].mean()

# Plot the average income by year
plt.figure(figsize=(12, 6))
average_income_by_year.plot(kind='line', color='orange')
plt.xlabel('Year')
plt.ylabel('Average Income')
plt.title('Average Income Over the Years')
plt.show()

# Additional insights: Population by profession for the latest year
latest_year = data_df['year'].max()
population_by_profession = data_df[data_df['year'] == latest_year].groupby('profession').size()

# Plot the population by profession for the latest year
plt.figure(figsize=(12, 6))
population_by_profession.plot(kind='bar')
plt.xlabel('Profession')
plt.ylabel('Population')
plt.title(f'Population by Profession in Year {latest_year}')
plt.show()
```
