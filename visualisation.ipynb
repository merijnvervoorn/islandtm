```python
import csv
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Read the data from the CSV file
gdp = pd.read_csv('data_all_teams_gdp.csv')

# Create a new figure
plt.figure()

# Plot the data
plt.plot(gdp['year'], gdp['gdp'], label='GDP')

```

```python
# Read the data from the CSV file into a DataFrame
data_df = pd.read_csv('data_all_teams.csv')

# Plot the data
plt.figure(figsize=(10, 6))
data_df.groupby('profession')['income'].mean().plot(kind='bar')
plt.xlabel('Profession')
plt.ylabel('Average Income')
plt.title('Average Income by Profession')
plt.show()
```
