```python
import csv
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Read the data from the CSV file
gdp = pd.read_csv('data_all_teams_gdp.csv')

# Create a new figure
plt.figure()

# Plot the data
plt.plot(gdp['year'], gdp['gdp'], label='GDP')

```

```python
# Read the data from the CSV file into a DataFrame
data_df = pd.read_csv('data_all_teams.csv')

# Plot the data
plt.figure(figsize=(10, 6))
data_df.groupby('profession')['income'].mean().plot(kind='bar')
plt.xlabel('Profession')
plt.ylabel('Average Income')
plt.title('Average Income by Profession')
plt.show()
```

```python
# Group by year and count the number of citizens
citizen_population = data_df.groupby('year').size()

# Plot the citizen population for every year
plt.figure(figsize=(12, 6))
citizen_population.plot(kind='line')
plt.xlabel('Year')
plt.ylabel('Citizen Population')
plt.title('Citizen Population Over the Years')
plt.show()

# Additional insights: Average income by year
average_income_by_year = data_df.groupby('year')['income'].mean()

# Plot the average income by year
plt.figure(figsize=(12, 6))
average_income_by_year.plot(kind='line', color='orange')
plt.xlabel('Year')
plt.ylabel('Average Income')
plt.title('Average Income Over the Years')
plt.show()

# Additional insights: Population by profession for the latest year
latest_year = data_df['year'].max()
population_by_profession = data_df[data_df['year'] == latest_year].groupby('profession').size()

# Plot the population by profession for the latest year
plt.figure(figsize=(12, 6))
population_by_profession.plot(kind='bar')
plt.xlabel('Profession')
plt.ylabel('Population')
plt.title(f'Population by Profession in Year {latest_year}')
plt.show()
```

```python
# Filter out the rows where the profession is 'child'
filtered_data_df = data_df[data_df['profession'] != 'child']

# Group by year and profession, and calculate the mean income
average_income_per_profession = filtered_data_df.groupby(['year', 'profession'])['income'].mean().unstack()

# Plot the average income per profession over the years
plt.figure(figsize=(14, 8))
average_income_per_profession.plot(ax=plt.gca())
plt.xlabel('Year')
plt.ylabel('Average Income')
plt.title('Average Evolution of Income per Profession (Excluding Children)')
plt.legend(title='Profession')
plt.show()
```

```python
# Filter out the rows where the profession is 'child'
filtered_data_df = data_df[data_df['profession'] != 'child']

# Group by year and profession, and count the number of citizens
citizens_per_profession_per_year = filtered_data_df.groupby(['year', 'profession']).size().unstack()

# Plot the number of citizens per profession over the years
plt.figure(figsize=(14, 8))
citizens_per_profession_per_year.plot(ax=plt.gca())
plt.xlabel('Year')
plt.ylabel('Number of Citizens')
plt.title('Number of Citizens per Profession Over the Years (Excluding Children)')
plt.legend(title='Profession')
plt.show()
```

```python
# Filter the data for children
children_data = data_df[data_df['profession'] == 'child']

# Group by year and count the number of children
children_population = children_data.groupby('year').size()

# Plot the children population over the years
plt.figure(figsize=(12, 6))
children_population.plot(kind='line', color='blue')
plt.xlabel('Year')
plt.ylabel('Number of Children')
plt.title('Number of Children Over the Years')
plt.show()
```

```python
# Plot the GDP over the years
plt.figure(figsize=(12, 6))
plt.plot(gdp['year'], gdp['gdp'], label='GDP', color='green')
plt.xlabel('Year')
plt.ylabel('GDP')
plt.title('GDP Over the Years')
plt.legend()
plt.show()

# Plot the average income by gender over the years
average_income_by_gender = data_df.groupby(['year', 'sex'])['income'].mean().unstack()
plt.figure(figsize=(12, 6))
average_income_by_gender.plot(ax=plt.gca())
plt.xlabel('Year')
plt.ylabel('Average Income')
plt.title('Average Income by Gender Over the Years')
plt.legend(title='Gender')
plt.show()

# Plot the population distribution by gender for the latest year
population_by_gender_latest_year = data_df[data_df['year'] == latest_year].groupby('sex').size()
plt.figure(figsize=(8, 6))
population_by_gender_latest_year.plot(kind='bar', color=['blue', 'orange', 'green'])
plt.xlabel('Gender')
plt.ylabel('Population')
plt.title(f'Population Distribution by Gender in Year {latest_year}')
plt.show()
```

```python
# Filter the data for 'X' gender
x_gender_data = data_df[data_df['sex'] == 'X']

# Group by year and count the number of 'X' gender
x_gender_population = x_gender_data.groupby('year').size()

# Plot the 'X' gender population over the years
plt.figure(figsize=(12, 6))
x_gender_population.plot(kind='line', color='purple')
plt.xlabel('Year')
plt.ylabel('Number of X Gender')
plt.title('Number of X Gender Over the Years')
plt.show()
# Filter the data for 'F' gender
f_gender_data = data_df[data_df['sex'] == 'F']

# Group by year and count the number of 'F' gender
f_gender_population = f_gender_data.groupby('year').size()

# Plot the 'F' gender population over the years
plt.figure(figsize=(12, 6))
f_gender_population.plot(kind='line', color='red')
plt.xlabel('Year')
plt.ylabel('Number of F Gender')
plt.title('Number of F Gender Over the Years')
plt.show()

# Filter the data for 'M' gender
m_gender_data = data_df[data_df['sex'] == 'M']

# Group by year and count the number of 'M' gender
m_gender_population = m_gender_data.groupby('year').size()

# Plot the 'M' gender population over the years
plt.figure(figsize=(12, 6))
m_gender_population.plot(kind='line', color='blue')
plt.xlabel('Year')
plt.ylabel('Number of M Gender')
plt.title('Number of M Gender Over the Years')
plt.show()
```

```python
# Group by profession and gender, and count the number of citizens
gender_per_profession = data_df.groupby(['profession', 'sex']).size().unstack()

# Plot the gender distribution per profession
plt.figure(figsize=(14, 8))
gender_per_profession.plot(kind='bar', stacked=True, ax=plt.gca())
plt.xlabel('Profession')
plt.ylabel('Number of Citizens')
plt.title('Gender Distribution per Profession')
plt.legend(title='Gender')
plt.show()
```

```python
# Filter out the rows where the profession is 'child'
filtered_data_df = data_df[data_df['profession'] != 'child']

# Plot the distribution of income excluding children per year
years = filtered_data_df['year'].unique()
plt.figure(figsize=(14, 10))

for year in years:
    plt.hist(filtered_data_df[filtered_data_df['year'] == year]['income'], bins=50, alpha=0.5, label=f'Year {year}')

plt.xlabel('Income')
plt.ylabel('Frequency')
plt.title('Distribution of Income (Excluding Children) Per Year')
plt.legend()
plt.show()
```

```python
# Read the updated data from the CSV file into a DataFrame
data_df = pd.read_csv('data_all_teams.csv')

# Plot the average income by age
plt.figure(figsize=(10, 6))
data_df.groupby('age')['income'].mean().plot(kind='bar')
plt.xlabel('Age')
plt.ylabel('Average Income')
plt.title('Average Income by Age')
plt.show()

# Plot the population distribution by age
plt.figure(figsize=(10, 6))
data_df['age'].value_counts().sort_index().plot(kind='bar')
plt.xlabel('Age')
plt.ylabel('Population')
plt.title('Population Distribution by Age')
plt.show()

# Plot the average income by age and gender
average_income_by_age_gender = data_df.groupby(['age', 'sex'])['income'].mean().unstack()
plt.figure(figsize=(12, 6))
average_income_by_age_gender.plot(kind='line', ax=plt.gca())
plt.xlabel('Age')
plt.ylabel('Average Income')
plt.title('Average Income by Age and Gender')
plt.legend(title='Gender')
plt.show()
```

```python
# Filter the data for the year 100
data_year_100 = data_df[data_df['year'] == 100]

# Plot the population distribution by profession for the year 100
plt.figure(figsize=(12, 6))
data_year_100['profession'].value_counts().plot(kind='bar')
plt.xlabel('Profession')
plt.ylabel('Population')
plt.title('Population Distribution by Profession in Year 100')
plt.show()

# Plot the population distribution by gender for the year 100
plt.figure(figsize=(8, 6))
data_year_100['sex'].value_counts().plot(kind='bar', color=['blue', 'orange', 'green'])
plt.xlabel('Gender')
plt.ylabel('Population')
plt.title('Population Distribution by Gender in Year 100')
plt.show()

# Plot the average income by profession for the year 100
plt.figure(figsize=(12, 6))
data_year_100.groupby('profession')['income'].mean().plot(kind='bar')
plt.xlabel('Profession')
plt.ylabel('Average Income')
plt.title('Average Income by Profession in Year 100')
plt.show()
# Plot the average income by age for the year 100
plt.figure(figsize=(12, 6))
data_year_100.groupby('age')['income'].mean().plot(kind='bar')
plt.xlabel('Age')
plt.ylabel('Average Income')
plt.title('Average Income by Age in Year 100')
plt.show()
```

```python
# Filter the data for fishers and farmers
fishers_farmers_data = data_df[data_df['profession'].isin(['fisher', 'farmer'])]

# Get the last 10 years
last_10_years = fishers_farmers_data['year'].unique()[-10:]

# Filter the data for the last 10 years
last_10_years_data = fishers_farmers_data[fishers_farmers_data['year'].isin(last_10_years)]

# Print the last 10 years of income of fishers and farmers
print(last_10_years_data[['year', 'profession', 'income']])
# Calculate the average income for fishers and farmers over the last 10 years
average_income_fishers_farmers = last_10_years_data.groupby(['year', 'profession'])['income'].mean().unstack()

# Print the average income for fishers and farmers over the last 10 years
print(average_income_fishers_farmers)
```

```python
# Filter out the rows where the profession is 'child'
filtered_data_df = data_df[data_df['profession'] != 'child']

# Group by year and profession, and calculate the mean income
average_income_per_profession = filtered_data_df.groupby(['year', 'profession'])['income'].mean().unstack()

# Plot the average income per profession over the years in sections of 10 years
years = range(0, 101, 10)
for start_year in years:
    end_year = start_year + 10
    plt.figure(figsize=(14, 8))
    average_income_per_profession.loc[start_year:end_year].plot(ax=plt.gca())
    plt.xlabel('Year')
    plt.ylabel('Average Income')
    plt.title(f'Average Evolution of Income per Profession (Excluding Children) from Year {start_year} to {end_year}')
    plt.legend(title='Profession')
    plt.show()
```
